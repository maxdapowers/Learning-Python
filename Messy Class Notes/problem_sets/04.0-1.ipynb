```python

import csv

fn = '../data/titanic.csv'

with open(fn,encoding='utf-8') as f:
    data = [{k: str(v) for k, v in row.items()}
        for row in csv.DictReader(f, skipinitialspace=True)]
#use the code above to load  data into a list
#    of dictionaries for each problem below
```

```python
'''1
Print the names of all the fields in the format:
field1
field2
field3
'''

for field in data[0].keys():
    print(field)

```

```python
'''2
What is the total number of male survivor passengers in the dataset?
'''

n = 0 # create the counting variable
for row in data:
    if 'Male' in row['Gender']:
        n += 1 #iterate the counting variable for each time the if statement is true
print(n)
```

```python
'''3
What is the total number of female survivors in the dataset?
'''

n = 0 # create the counting variable
for row in data:
    if 'Female' in row['Gender']:
        if 'Survivor' in row['Status']:
            n += 1 #iterate the counting variable for each time the if statement is true
print(n)
```

```python
'''4
Print a list of all the passengers who were from the United States
'''
ppl = []
for row in data:
    if 'United States' in row['Home Country']:
        ppl.append(row['Titanic passager'])
print(ppl)
```

```python
'''5
What is the total number of passengers from the United States, England or Canada?
(provide 1 total)
'''

ppl = 0

for row in data:
    if 'United States' in row['Home Country']:
        ppl += 1
    if 'England' in row['Home Country']:
        ppl += 1
    if 'Canada' in row['Home Country']:
        ppl += 1
        
print(ppl)
```

```python
'''6
Create a list of all of the occupations for people in the dataset.
Give a count of the number of people with each occupation.
'''

jobs = {} # creates a dictionary
for row in data:
    k = row['Job']
    if k in jobs: # checks to see if the job is already in the dictionary's keys
        jobs[k] += 1 # increments the value (from assigned the the key k)
    else:
        jobs[k] = 1 # if the channel isn't already in the dictionaries keys, create it and set it to one
        
# print(jobs)
for job,count in jobs.items():
    print(job,count) 
```

```python
def fareItems(items):
    d = {'#':0,'s':0,'d':0}
    if len(items) == 1:
        d['#'] = int(items[0].replace('£','')) # .replace() removes the £
    elif len(items) == 2:
        d['#'] = int(items[0].replace('£','')) 
        d['s'] = int(items[0].replace('s','')) 
    elif len(items) == 3:
        d['#'] = int(items[0].replace('£','')) 
        d['s'] = int(items[0].replace('s','')) 
        d['d'] = int(items[0].replace('d',''))  
    return d

l = ['£1']
print(fareItems(l))

l = ['£1', '10s']
print(fareItems(l))
```
