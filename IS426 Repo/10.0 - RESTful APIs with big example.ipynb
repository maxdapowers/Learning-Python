```python
# https://openskynetwork.github.io/opensky-api/rest.html

url = '''https://opensky-network.org/api/states/all?
    lamin=44&lomin=-75.7&lamax=45&lomax=-74'''
r = requests.get(url)

#print r.headers
print(r.text)
# [[lat1,lon1],[lat2,lon2]]
adata = json.loads(r.text)

```

```python
points = []
for state in adata['states']:
    points.append(state[5:7])
print(points)
```

```python
from ipyleaflet import Map, basemaps, basemap_to_tiles,CircleMarker

m = Map(
    basemap=basemap_to_tiles(basemaps.OpenStreetMap.Mapnik),
    center=(45, -76),
    zoom=3
    )
for point in points:
    circle_marker = CircleMarker()
    circle_marker.location = (point[1],point[0])
    circle_marker.radius = 1
    circle_marker.color = "red"
    circle_marker.fill_color = "red"

    m.add_layer(circle_marker)
m
```

```python
#################################### THIS HAS EVERYTHING FROM THE SEMESTER ######################################

# Create a CSV of 5 zipcodes.  Create a script which iterates over it a looks up the temperture and conditions
#   for each location.  The script should store the results in a mysql table with the following fields: 
# #    datetime,zip,location_name,tempf,conditions,json
# the json field should be the raw json response from the api.

import csv,requests,json
import pymysql
import secrets
import datetime

conn = pymysql.connect(host=secrets.host, port=3306, user=secrets.user,
                       passwd=secrets.passwd, db='is426', autocommit=True)

cur = conn.cursor(pymysql.cursors.DictCursor)

sql = '''DROP TABLE IF EXISTS `conlontj_weather`;'''
cur.execute(sql)


sql = '''
CREATE TABLE IF NOT EXISTS `conlontj_weather` (
  `id` int(5) NOT NULL AUTO_INCREMENT,
  `datetime` datetime NOT NULL,
  `zip` varchar(20) NOT NULL,
  `location_name` varchar(200) NOT NULL,
  `tempf` varchar(5) NOT NULL,
  `conditions` varchar(200) NOT NULL,
  `json` varchar(500) NOT NULL
  PRIMARY KEY (`id`)
) ENGINE=MyISAM DEFAULT CHARSET=latin1 AUTO_INCREMENT=1 ;
'''
cur.execute(sql)

csv_str = '''zipcode
13676
00000
08816
01220
12061'''
f = open("zips.csv","w")
f.write(csv_str)
f.close()
with open('zips.csv') as f:
    data = [{k: str(v) for k, v in row.items()}
        for row in csv.DictReader(f, skipinitialspace=True)]

sql ='''INSERT INTO conlontj_weather (`datetime`,`zip`,`location_name`,`tempf`,`conditions`,`json`)
VALUES
(%s,%s,%s,%s,%s,%s);
'''

for row in data:
    z = row['zipcode']
    url = f'https://api.openweathermap.org/data/2.5/weather?zip={z},us&APPID=a6962678a5cba51e8db12b46bc87a867'
    print(z)
    r = requests.get(url)
    print(r.text)
    resp = json.loads(r.text)
    location = resp['name']
    tempf = round((resp['main']['temp'] - 273.15) * 9/5 + 32.0,1)
    cl = []
    for cond in resp['weather']:
        cl.append(cond['main'])
    cs = ','.join(cl)
    print(tempf,cs)
    dts = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S") # turns the present date and time from an object to a string
    cur.execute(sql,(dts,z,location,tempf,cs,r.text)) 
    break


```
